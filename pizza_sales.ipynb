```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px # Data visualization library  easy aur quick plots
```

```python
# import the data

pizza_sales=pd.read_csv(r"C:\Users\nimma\Downloads\pizza_sales.csv",encoding_errors='ignore')
pizza_sales.head(10)
```

```python
# information the data

pizza_sales.info()
```

```python
# find the null values

pizza_sales.isnull().sum()
```

```python
# find the duplicate

pizza_sales.duplicated().sum()
```

```python
pizza_sales['pizza_size'].unique()
```

```python
# change the specific rows in pizza size

pizza_sales['pizza_size']=pizza_sales['pizza_size'].replace({'s' :'Regural','M':'Medium','L':'Large','XL':'X-Large','XXL':'XX-Large'})

pizza_sales.head(10)
```

```python
# convert datetime format day_mpnth_year change to year_month_day

pizza_sales['order_date']=pd.to_datetime(pizza_sales['order_date'], format=('%d-%m-%Y'))

# Format as Year-Month-Day

pizza_sales['order_date']=pizza_sales['order_date'].dt.strftime('%Y-%m-%d')
```

```python
# convert data columns

pizza_sales['order_date']=pd.to_datetime(pizza_sales['order_date'])
```

```python
# new column in day and month

pizza_sales['order_day']=pizza_sales['order_date'].dt.day_name()
pizza_sales['order_month']=pizza_sales['order_date'].dt.month_name()


pizza_sales.head(5)
```

```python
# total revenue

total_revenue=pizza_sales['total_price'].sum()

print(total_revenue)
```

```python
# Average order value

avearge_order_value=pizza_sales['total_price'].sum()/pizza_sales['order_id'].nunique()

print(avearge_order_value)
```

```python
# total pizza sold
total_pizza_sold=pizza_sales['quantity'].sum()

print(total_pizza_sold)
```

```python
# Total order

total_orders=pizza_sales['order_id'].nunique()

print(total_orders)
```

```python
# average pizzas per order

avg_pizza_per_order=pizza_sales['quantity'].sum()/pizza_sales['order_id'].nunique()

print(avg_pizza_per_order)
```

```python
pizza_sales['order_month'].isnull().sum()
```

```python
# daily trend for total_orders

daily_total_order=pizza_sales.groupby(['order_day'],as_index=False)['total_price'].sum().sort_values(by='total_price',ascending=True)

sns.barplot(data=daily_total_order,x='order_day',y='total_price',hue='order_day')
plt.xticks(rotation=45,ha='right')
plt.title('daily for total orders')
```

```python
# monthly trends off total orders


monthly_total_order=pizza_sales.groupby(['order_month'],as_index=False)['total_price'].sum().sort_values(by='total_price',ascending=True)

plt.figure(figsize=(10,4))
sns.lineplot(data=monthly_total_order,x='order_month',y='total_price')
plt.xticks(rotation=45)
plt.title('Monthly trends for total orders')
```

```python
# Percentage of sales by pizza category

sales_category=pizza_sales.groupby(['pizza_category'])['total_price'].sum().reset_index()

'''plt.pie(sales ,labels=sales.index,autopct='%1.1f%%')
plt.title(' percentage sales by pizza category')
plt.show()'''
fig = px.pie(sales_category, 
             values='total_price', 
             names='pizza_category', 
             hole=0.5, 
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by pizza Category', title_font=dict(size=24))

fig.show()
```

```python
# percentage of sales by pizza size

pizza_size_sales=pizza_sales.groupby(['pizza_size'])['total_price'].sum().reset_index()


fig = px.pie(pizza_size_sales, 
             values='total_price', 
             names='pizza_size', 
             hole=0.4, 
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by pizza size', title_font=dict(size=24))

fig.show()
```

```python
# pizza sold by pizza category

pizza_sold_by_pizza_categort=pizza_sales.groupby(['pizza_category'],as_index=False)['quantity'].sum().sort_values(by='quantity',ascending=False)

sns.barplot(data=pizza_sold_by_pizza_categort,x='pizza_category',y='quantity',hue='pizza_category')
plt.title('total pizza sold by pizza category')
plt.xlabel('pizza_category')
plt.ylabel('total quantity')
```

```python
# top 5 pizzas by in revenue

pizza_sales

best_seller_revenue=pizza_sales.groupby(['pizza_name'],as_index=False)['total_price'].sum().sort_values(by='total_price',ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(data=best_seller_revenue,x='pizza_name',y='total_price',hue='pizza_name')
plt.xticks(rotation=45,ha='right')
plt.title('top 5 pizzas by in revenue')
plt.xlabel('pizza_name')
plt.ylabel('total amount')
plt.tight_layout()
plt.show()
```

```python
# top 5 pizzas by quantity


best_seller_quantity=pizza_sales.groupby(['pizza_name'],as_index=False)['quantity'].sum().sort_values(by='quantity',ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(data=best_seller_quantity,x='pizza_name',y='quantity',hue='pizza_name')
plt.xticks(rotation=45,ha='right')
plt.title('top 5 pizzas by quantity')
plt.xlabel('pizza_name')
plt.ylabel(' total quantity ')
plt.tight_layout()
plt.show()
```

```python

# top 5 pizzas total orders

pizza_sales

best_seller_total_orders=pizza_sales.groupby(['pizza_name'],as_index=False)['order_id'].nunique().sort_values(by='order_id',ascending=False).head(5)

sns.barplot(data=best_seller_total_orders,x='pizza_name',y='order_id',hue='pizza_name')
plt.xticks(rotation=45,ha='right')
plt.title('top 5 best sellers in total orders')
plt.xlabel('pizza_name')
plt.ylabel('total orders')
plt.tight_layout()
plt.show()
```

```python

# bottom 5 pizzas in revenu

best_seller_revenue=pizza_sales.groupby(['pizza_name'],as_index=False)['total_price'].sum().sort_values(by='total_price',ascending=True).head(5)

plt.figure(figsize=(10,6))
sns.barplot(data=best_seller_revenue,x='pizza_name',y='total_price',hue='pizza_name')
plt.xticks(rotation=45,ha='right')
plt.title('bottom 5 pizzas revenue')
plt.xlabel('pizza_name')
plt.ylabel('total amount')
plt.tight_layout()
plt.show()
```

```python
# bottom 5 pizzas by quantity


best_seller_quantity=pizza_sales.groupby(['pizza_name'],as_index=False)['quantity'].sum().sort_values(by='quantity',ascending=True).head(5)

plt.figure(figsize=(10,6))
sns.barplot(data=best_seller_quantity,x='pizza_name',y='quantity',hue='pizza_name')
plt.xticks(rotation=45,ha='right')
plt.title('Bottom 5 pizzas by quantity')
plt.xlabel('pizza_name')
plt.ylabel(' total quantity ')
plt.show()
```

```python
# bottom 5 pizzas in total orders

best_seller_total_orders=pizza_sales.groupby(['pizza_name'],as_index=False)['order_id'].nunique().sort_values(by='order_id',ascending=True).head(5)

plt.figure(figsize=(10,6))
sns.barplot(data=best_seller_total_orders,x='pizza_name',y='order_id',hue='pizza_name')
plt.xticks(rotation=45,ha='right')
plt.title('bottom 5 pizzas by  total orders')
plt.xlabel('pizza_name')
plt.ylabel('total orders')
plt.show()
```
